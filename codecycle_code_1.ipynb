```python
# Extract the subheadings and paragraphs using the appropriate HTML tags
subheadings = []
paragraphs = []

# Find all the step elements
steps = soup.find_all('div', {'class': 'step'})  # Adjust this class to match your HTML structure

for step in steps:
    # Extract subheading (if exists)
    subheading_element = step.find('b')  # Assuming the subheading is in a <b> tag
    if subheading_element is not None:
        subheading_text = subheading_element.text.strip().replace('\n', ' ')
        subheading_text = subheading_text.encode('ascii', errors='ignore').decode()
        subheading_text = re.sub(r'[^\w\s]', '', subheading_text)  # Remove unwanted characters
        print(f"Subheading: {subheading_text}")
        subheadings.append(subheading_text)

        # Remove the subheading from the step to clean the paragraph text later
        subheading_element.extract()

    # Remove any span tags within the step
    for span_tag in step.find_all('span'):
        span_tag.extract()

    # Extract the paragraph text
    paragraph_text = step.text.strip().replace('\n', ' ').replace('  ', ' ')
    paragraph_text = paragraph_text.encode('ascii', errors='ignore').decode()
    paragraph_text = re.sub(r'[^\w\s]', '', paragraph_text)  # Remove unwanted characters
    print(f"Paragraph: {paragraph_text}")
    paragraphs.append(paragraph_text)
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import re

# Create a CSV file in Colab environment
output_file = '/content/wikihow.csv'

# Number of articles to scrape
for count in range(5):  # Adjust the range for your needs

    # URL of WikiHow random page
    url = 'https://www.wikihow.com/Special:Randomizer'

    # Send an HTTP request to the URL and receive the HTML content
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the article title
    article_title = soup.find('title').text.strip()

    print(article_title + " " + str(count))

    # Initialize lists to store subheadings and paragraphs
    subheadings = []
    paragraphs = []

    # Find all steps in the article
    steps = soup.find_all('div', {'class': 'step'})

    for step in steps:
        # Extract subheading (if present) and clean it
        subheading_element = step.find("b")
        if subheading_element is not None:
            subheading_text = subheading_element.text.strip().replace('\n', '')
            subheading_text = subheading_text.encode('ascii', errors='ignore').decode()
            subheading_text = re.sub(",", "", subheading_text)
            subheadings.append(subheading_text)

            # Remove the subheading element to avoid duplication in paragraph extraction
            subheading_element.extract()

        # Remove any span tags to clean up the text
        for span_tag in step.find_all('span'):
            span_tag.extract()

        # Extract and clean the paragraph text
        paragraph_text = step.text.strip().replace('\n', ' ')
        paragraph_text = paragraph_text.encode('ascii', errors='ignore').decode()
        paragraph_text = re.sub(",", "", paragraph_text)
        paragraphs.append(paragraph_text)

    # Save the data to a CSV file
    if len(subheadings):
        with open(output_file, mode='a', newline='', encoding="utf-8") as csv_file:
            writer = csv.writer(csv_file)
            for i in range(len(subheadings)):
                writer.writerow([article_title, subheadings[i], paragraphs[i]])

# Verify the file exists
print("CSV file saved at:", output_file)
```

```python
# Max input/output lengths
max_input_length = 1024
max_output_length = 64

def process_data_to_model_inputs(batch, tokenizer):
    # Tokenize the inputs (e.g., paragraphs) and labels (e.g., headings)
    inputs = tokenizer(
        batch["paragraph"],   # Tokenizing the paragraphs
        padding="max_length", # Pad to max length
        truncation=True,      # Truncate if longer than max_input_length
        max_length=max_input_length
    )

    outputs = tokenizer(
        batch["heading"],     # Tokenizing the headings
        padding="max_length", # Pad to max length
        truncation=True,      # Truncate if longer than max_output_length
        max_length=max_output_length
    )

    # Set the input IDs and attention masks
    batch["input_ids"] = inputs.input_ids
    batch["attention_mask"] = inputs.attention_mask

    # Create global attention masks
    # Initialize global_attention_mask with zeros for each input
    batch["global_attention_mask"] = [
        [0 for _ in range(len(batch["input_ids"][0]))] for _ in range(len(batch["input_ids"]))
    ]

    # Set the first token's global attention mask to 1
    # This typically means the [CLS] token gets global attention
    for i in range(len(batch["global_attention_mask"])):
        batch["global_attention_mask"][i][0] = 1

    # Set labels from the output input_ids
    batch["labels"] = outputs.input_ids

    # Ignore padding tokens in the labels by setting them to -100
    batch["labels"] = [
        [-100 if token == tokenizer.pad_token_id else token for token in labels]
        for labels in batch["labels"]
    ]

    return batch
```

```python
from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainer, Seq2SeqTrainingArguments, AutoTokenizer
from datasets import load_metric
import transformers
import os

# Load the model and tokenizer
model_name = "allenai/led-base-16384"
led = AutoModelForSeq2SeqLM.from_pretrained(model_name, gradient_checkpointing=True, use_cache=False)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Set configuration
led.config.num_beams = 2
led.config.max_length = 64
led.config.min_length = 2
led.config.length_penalty = 2.0
led.config.early_stopping = True
led.config.no_repeat_ngram_size = 3

# Load the ROUGE metric
rouge = load_metric("rouge")

# Compute metrics function
def compute_metrics(pred):
    labels_ids = pred.label_ids
    pred_ids = pred.predictions

    # Convert IDs to strings
    pred_str = tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
    labels_ids[labels_ids == -100] = tokenizer.pad_token_id
    label_str = tokenizer.batch_decode(labels_ids, skip_special_tokens=True)

    # Compute ROUGE scores
    rouge_output = rouge.compute(
        predictions=pred_str,
        references=label_str,
        rouge_types=["rouge2"]
    )["rouge2"].mid

    return {
        "rouge2_precision": round(rouge_output.precision, 4),
        "rouge2_recall": round(rouge_output.recall, 4),
        "rouge2_fmeasure": round(rouge_output.fmeasure, 4)
    }

# Define batch size
batch_size = 16  # Set the batch size

# Define training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="/content/results/checkpoint",  # Directory to save model outputs
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    logging_steps=5,
    eval_steps=10,
    save_steps=500,  # Adjust this based on your training
    save_total_limit=2,
    gradient_accumulation_steps=4,
    num_train_epochs=10,
    predict_with_generate=True
)

# Set logging verbosity
transformers.logging.set_verbosity_info()

# Ensure the directory exists
os.makedirs("/content/results/checkpoint", exist_ok=True)

# Initialize the Seq2SeqTrainer
trainer = Seq2SeqTrainer(
    model=led,
    args=training_args,
    compute_metrics=compute_metrics,
    train_dataset=train_dataset,  # Ensure train_dataset is properly loaded
    eval_dataset=val_dataset,    # Ensure val_dataset is properly loaded
    tokenizer=tokenizer
)
# Save the model and tokenizer after training
led.save_pretrained("/content/results/checkpoint")
tokenizer.save_pretrained("/content/results/checkpoint")

# Start training
trainer.train()

# Save the model and tokenizer after training
led.save_pretrained("/content/results/checkpoint")
tokenizer.save_pretrained("/content/results/checkpoint")
```

```python
from transformers import LEDTokenizer, LEDForConditionalGeneration, pipeline
from datasets import Dataset
import torch
from keybert import KeyBERT  # You will need to install KeyBERT (pip install keybert)

# Load tokenizer and model
tokenizer = LEDTokenizer.from_pretrained("/content/results/checkpoint")
model = LEDForConditionalGeneration.from_pretrained("/content/results/checkpoint")

# Load a separate summarization pipeline (optional, use same model for summarization if needed)
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Initialize KeyBERT for tag/keyword extraction
kw_model = KeyBERT()

# Move model to device
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def generate_answer(batch):
    # Tokenize the input text (paragraph)
    inputs = tokenizer(batch["paragraph"], padding="max_length", max_length=512, return_tensors="pt", truncation=True)

    input_ids = inputs.input_ids.to(device)
    attention_mask = inputs.attention_mask.to(device)

    # Create global attention mask
    global_attention_mask = torch.zeros_like(attention_mask)
    global_attention_mask[:, 0] = 1  # Set global attention on the first token

    # Generate the heading
    predicted_heading_ids = model.generate(
        input_ids=input_ids,
        attention_mask=attention_mask,
        global_attention_mask=global_attention_mask,
        max_length=64,
        min_length=10,
        num_beams=2,
        length_penalty=2.0,
        early_stopping=True,
        no_repeat_ngram_size=3
    )

    # Decode the heading
    batch["generated_heading"] = tokenizer.batch_decode(predicted_heading_ids, skip_special_tokens=True)[0]

    # Summarization using BART or the same LED model
    summarized_text = summarizer(batch["paragraph"], max_length=50, min_length=25, do_sample=False)
    batch["summary"] = summarized_text[0]['summary_text']

    # Extract tags/keywords using KeyBERT
    tags = kw_model.extract_keywords(batch["paragraph"], keyphrase_ngram_range=(1, 2), stop_words='english', top_n=5)
    batch["tags"] = [tag[0] for tag in tags]

    return batch

# Assuming `test` is your dataset (ensure it's a datasets.Dataset object)
if isinstance(test, pd.DataFrame):
    test = Dataset.from_pandas(test)

# Apply the function
result = test.map(generate_answer, batched=True, batch_size=2)

# Print the output
for res in result:
    print(f"Heading: {res['generated_heading']}")
    print(f"Summary: {res['summary']}")
    print(f"Tags: {', '.join(res['tags'])}")
    print("\n")
```
